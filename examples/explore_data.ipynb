### Plot the trace (8 channels) with its corresponding -true- segmentation
Notice that the "BG" class indicates the presence of an event (when 0) or not (when 1). Only the corresponding class of the event is activated (value of 1) when the event is present, and turns back to 0 when it is no longer there.

Volcano data examples for each dataset are shown below, displaying two for each class, one with the least amount of stations and another with the most amount of stations available. Note that signals are standardized by dividing the whole trace by the maximum absolute value accross stations and samples.

```python
for i in range(len(NVCh_data)):
    example_trace = np.load(NVCh_data[i])
    data_utils.plot_trace(example_trace,
                            n_stations=8,
                            save_path=None,
                            save=False,
                            title=NVCh_data[i][8:-6],
                            num=i,
                            dpi=100,
                            figsize=(10, 6),
                            font_size=9)
```
